```
# 01_eda.ipynb (Complete EDA)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option("display.max_columns", 50)

claims = pd.read_csv("../data/claims.csv")
providers = pd.read_csv("../data/providers.csv")
patients = pd.read_csv("../data/patients.csv")

print("Data loaded successfully!")
print("Claims shape:", claims.shape)
print("Providers shape:", providers.shape)
print("Patients shape:", patients.shape)

display(claims.head())
display(providers.head())
display(patients.head())

claims.info()
providers.info()
patients.info()

print(claims.isna().sum())
print(providers.isna().sum())
print(patients.isna().sum())

fraud_counts = claims["is_fraud"].value_counts(normalize=True)
print(fraud_counts)
claims["is_fraud"].value_counts().plot(kind="bar", title="Fraud Count")
plt.show()

plt.hist(claims["claim_amount"], bins=50)
plt.title("Claim Amount Distribution")
plt.show()

display(claims["diagnosis_code"].value_counts().head(10))
display(claims["procedure_code"].value_counts().head(10))

provider_fraud = claims.groupby("provider_id")["is_fraud"].mean().sort_values(ascending=False)
display(provider_fraud.head(10))

claims_prov = claims.merge(providers, on="provider_id", how="left")
display(claims_prov.head())

claims["claim_date"] = pd.to_datetime(claims["claim_date"])
monthly = claims.set_index("claim_date").resample("M")["claim_id"].count()
monthly.plot(title="Claims Over Time")
plt.show()

display(patients["age"].describe())
plt.hist(patients["age"], bins=30)
plt.title("Age Distribution")
plt.show()

display(patients["gender"].value_counts())
patients["gender"].value_counts().plot(kind="bar")
plt.show()

display(patients["insurance_type"].value_counts())
patients["insurance_type"].value_counts().plot(kind="bar")
plt.show()

print("EDA Completed")
```
